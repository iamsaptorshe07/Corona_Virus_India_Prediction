```python
new_df = df[['Population Density per km2','Week','Attacked_People']]
```

```python
new_df
```

```python
new_df = new_df.rename(columns={"Population Density per km2": "PD"})
```

```python
from mpl_toolkits.mplot3d import Axes3D
# Visualizing multiple linear regression
%matplotlib inline
fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')
ax.scatter(new_df['PD'],new_df['Week'],new_df['Attacked_People'],color='red')
plt.show()
```

```python
from sklearn import linear_model
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(new_df[['PD','Week']],new_df[['Attacked_People']],test_size=0.10,random_state = 0)
print(x_test)
print(y_test)
```

```python
clf = linear_model.LinearRegression()
clf.fit(x_train,y_train)
clf.predict(x_test)
```

```python
clf.score(x_test,y_test)
```
